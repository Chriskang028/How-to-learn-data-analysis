```python
df_duplicate = df.drop_duplicates(subset='positionId', keep='first').copy()
df_duplicate.head()
```

```python
def get_Salary(string, method = 'bottom'):
    position = string.find('-')
    
    if position != -1:
        bottom_Salary = string[:position - 1]
        top_Salary = string[position + 1:len(string) - 1]
    else:
        bottom_Salary = string[:string.upper().find('K')]
        top_Salary = bottom_Salary
    
    if method == 'bottom':
        return bottom_Salary
    else:
        return top_Salary
        
# Cleaning the salary data with get_Salary function
df_duplicate['top_Salary'] = df_duplicate.salary.apply(get_Salary, method='top').astype('int')
df_duplicate['bottom_Salary'] = df_duplicate.salary.apply(get_Salary, method='bottom').astype('int')
```

```python
# Calculating the average salary, and put it into the column "Avg_Salary"
df_duplicate['Avg_Salary'] = df_duplicate.apply(lambda x:(x.top_Salary + x.bottom_Salary) / 2, axis=1)
```

```python
# Change data type
df_duplicate['city'] = df_duplicate.city.astype('category')
```
